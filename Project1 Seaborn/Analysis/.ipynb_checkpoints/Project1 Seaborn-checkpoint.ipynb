```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sorted_months_weekdays import *
from sort_dataframeby_monthorweek import *
import chart_studio.plotly as py
import plotly
import plotly.express as px
import plotly.io as pio
```

```python
data=pd.read_csv('hotel_bookings.csv')
data.head(2)
```

```python
data_copy=data.copy()
```

```python
data.shape
```

```python
data.info()
```

```python
#replace na values with the mean of the children
child_mean=data['children'].mean()
data['children']=data['children'].fillna(child_mean)
```

```python
data['children']=data['children'].astype('int64')
```

```python
#delete the rows that have zero value to adults , children and babies
data=data[~((data.adults == 0) & (data.children == 0) & (data.babies == 0))]
data.shape
```

```python
#Put the months in order
data = Sort_Dataframeby_Month(df=data,monthcolumnname='arrival_date_month')
data.head()
```

```python
#Sum the days and clients per reservation
data['members']=data['adults']+data['children']+data['babies']
data['total_days']=data['stays_in_weekend_nights']+data['stays_in_week_nights']
data.head(2)
```

# Country-Cancelations

```python
#choose only 3 columns from the original
country=data[['hotel' , 'country' , 'reservation_status']].copy()
country
```

```python
#choose only Canceled from the dataframe country
country_canc = country.loc[(country['reservation_status'] == 'Canceled')]
country_canc
```

```python
#choose only City Hotel from the country_canc dataframe , count the Canceled values , sort and take the top 5
city_canc=country_canc.loc[(country_canc['hotel']=='City Hotel')]
city_nation=city_canc.groupby('country').reservation_status.value_counts().sort_values(ascending=False).head().reset_index(name='Total1')
city_nation
```

```python
#choose only Resort Hotel from the country_canc dataframe , count the Canceled values , sort and take the top 5
resort_canc=country_canc.loc[(country_canc['hotel']=='Resort Hotel')]
resort_nation=resort_canc.groupby('country').reservation_status.value_counts().sort_values(ascending=False).head().reset_index(name='Total2')
resort_nation
```

```python
plt.subplot(1,2,1)
sns.set(rc={'figure.figsize':(10,8)})
h=sns.barplot(x='country' , y='Total2'  , palette='PuBuGn_r' , edgecolor='0.3' ,
            data=resort_nation ,  hatch='/' )

h.set_title('Resort Hotel' , fontsize=17 , weight='bold')
h.set_xlabel('Nationality' , fontsize=17 , weight='bold')
h.set_ylabel('Cancellations' , fontsize=17 , weight='bold')
h.tick_params(labelsize=12)

plt.subplot(1,2,2)
sns.set(rc={'figure.figsize':(10,7)})
h=sns.barplot(x='country' , y='Total1'  , palette='PuBuGn_r' , edgecolor='0.3' ,
            data=city_nation ,  hatch='/' )

h.set_title('City Hotel' , fontsize=17 , weight='bold')
h.set_xlabel('Nationality' , fontsize=17 , weight='bold')
h.set_ylabel('Cancellations' , fontsize=17 , weight='bold')
h.tick_params(labelsize=12)
```

# Accommodation type and Month period over year

```python
#Keep only Check-Out
dataset=data.loc[(data['reservation_status'] == 'Check-Out')]
dataset
```

```python
##choose only 2 columns from the dataset
c=dataset[[ 'arrival_date_month' , 'members']].copy()
```

```python
#Take only members<5
c1= c.loc[data['members'] <=5]
```

```python
mymap = {1:'Single', 2:'Double', 3:'Family', 4:'Family', 5:'Family'}
```

```python
c2=c1.applymap(lambda s: mymap.get(s) if s in mymap else s )
c2
```

```python
c3=c2.groupby(['arrival_date_month' , 'members'])['members'].count().reset_index(name='total')
c4=Sort_Dataframeby_Month(df=c3,monthcolumnname='arrival_date_month')
```

```python
#Normalization
c4_min = c4["total"].min()
c4_max = c4["total"].max()
c4["normalized"] = (c4["total"] - c4_min) / (c4_max - c4_min) * 100
c4.head()
```

```python
c4.columns=['arrival_date_month' , 'Accomodation Type' , 'total' , 'normalized']
```

```python
sns.set(rc={'figure.figsize':(15,7)})
f=sns.barplot(x='arrival_date_month' , y='normalized' , hue='Accomodation Type' , ci=None , palette='BuPu_r' , 
              edgecolor='0.3' , data=c4 )

f.set_title('Availability of Acomodation Type over time' , fontsize=17 , weight='bold')
f.set_xlabel('Months' , fontsize=17 , weight='bold')
f.set_ylabel('Checked Out Rooms(percentage)' , fontsize=17 , weight='bold')
f.tick_params(labelsize=12)

```
