```python
data=pd.read_csv('hotel_bookings.csv')
data.head(2)
```

```python
children_mean=data['children'].mean()
data['children']=data['children'].fillna(children_mean)
data['children']=data['children'].astype('int64')
```

```python
data = Sort_Dataframeby_Month(df=data,monthcolumnname='arrival_date_month')
data
```

```python
data['members']=data['adults']+data['children']+data['babies']
data['total_days']=data['stays_in_weekend_nights']+data['stays_in_week_nights']
data.head(2)
```

```python
#Convert object 'reservation_status_date' to datetime
data['reservation_status_date'] =  pd.to_datetime(data['reservation_status_date'], infer_datetime_format=True)
```

```python
#Create 2 new columns by the 'reservation_status_date' 
data['Year'] = data['reservation_status_date'].map(lambda x: x.year)
data['Month'] = data['reservation_status_date'].map(lambda x:x.month)
```
