```python
data=pd.read_csv(r'C:\Users\spana\Documents\Project1\HotelRevenueMgm\Project2\hotel_bookings.csv')
data.head()
```

```python
#Fill nan values in children with mean value
children_mean=data['children'].mean()
data['children']=data['children'].fillna(children_mean)
data['children']=data['children'].astype('int64')
```

```python
#Sort by Month
data = Sort_Dataframeby_Month(df=data,monthcolumnname='arrival_date_month')
```

```python
#Create 2 new columns with the sum of client & days
data['members']=data['adults']+data['children']+data['babies']
data['total_days']=data['stays_in_weekend_nights']+data['stays_in_week_nights']
data.head(2)
```

```python
#Convert object 'reservation_status_date' to datetime
data['reservation_status_date'] =  pd.to_datetime(data['reservation_status_date'], infer_datetime_format=True)
```

```python
#Create 2 new columns by the 'reservation_status_date' 
data['Year'] = data['reservation_status_date'].map(lambda x: x.year)
data['Month'] = data['reservation_status_date'].map(lambda x:x.month)
```

```python
url_coonections='https://assets1.lottiefiles.com/packages/lf20_5ohCYt.json'
url_reservations='https://assets9.lottiefiles.com/packages/lf20_ovhnyvxu.json'
url_canceled='https://assets10.lottiefiles.com/packages/lf20_mLtaRR.json'
url_noshow='https://assets8.lottiefiles.com/packages/lf20_L8eE99.json'
options = dict(loop=True, autoplay=True, rendererSettings=dict(preserveAspectRatio='xMidYMid slice'))

import dash_html_components as html
import dash_bootstrap_components as dbc
import plotly.express as px
app = JupyterDash(__name__, external_stylesheets=[dbc.themes.FLATLY])
app.layout = html.Div((
    

     dbc.Row([
        dbc.Col(html.H1("Hotel Revenue Managment Dashboard"),
                        style={'font-style': 'italic', 'color':'white'} ,
                        width={'size': 28, 'offset': 3},
                        )] , className='mb-5 mt-2'),

    
    dbc.Row([
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options, width="30%", height="30%", url=url_reservations)),
                dbc.CardBody([
                     html.H6('Reservations'),
                    html.H2(id='content-connections1', children="0")
                ])
            ]),
        ], width={'size':2 , 'offset': 2}),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options, width="30%", height="30%", url=url_coonections)),
                dbc.CardBody([
                    html.H6('Check Out'),
                    html.H2(id='content-connections', children="0")
                
                ])
            ]),
        ], width={'size':2 , 'offset': 0}),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options, width="30%", height="30%", url=url_canceled)),
                dbc.CardBody([
                    html.H6('Canceled'),
                    html.H2(id='content-connections2', children="0")
                ])
            ]),
        ],width={'size':2 , 'offset': 0}),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options, width="30%", height="30%", url=url_noshow)),
                dbc.CardBody([
                    html.H6('No Show'),
                    html.H2(id='content-connections3', children="0")
                ])
            ]),
        ],width={'size':2 , 'offset': 0})],className='mb-5'),
    


    dbc.Row([
     
    dbc.Col(
            dcc.RadioItems(
                    options=[
                            {'label': 'City Hotel', 'value': 'City Hotel'},
                       
                            {'label': 'Resort Hotel','value': 'Resort Hotel'} , 
                        {'label': 'All', 'value': 'All'},
                        
                                    ],
                         persistence=True ,
                         persistence_type='session',                                    
                         id='w_hoteltype' , 
                         value='City Hotel',
                         labelStyle={'display': 'inline-block'}),
                         width={'size': 5, "offset": 4})],className='mb-5 custom_size'),
        
    
    
        dbc.Row([
            
            dbc.Col(dcc.Dropdown(
        id='w_years', 
        value='2015', 
        options=[{'value': x, 'label': x} 
                 for x in ['2015', '2016', '2017']],
        clearable=False ,
        optionHeight=25 ,
        persistence=True ,
        persistence_type='session',
         style={'color':'black' , 'background-color': 'rgb(218,165,27)' , 'font-weight': 'bold'} ,
    ),width={'size': 2, "offset": 2 , 'order':1}),
            
            dbc.Col(dcc.Dropdown( 
    id='w_month', 
    value='January', 
    options=[{'value': value, 'label': value} 
                for value in month.values()],
             clearable=False ,
             optionHeight=25 ,
             persistence=True ,
             persistence_type='session',
             style={'color':'black' , 'background-color': 'rgb(218,165,27)' , 'font-weight': 'bold'}
    ),width={'size': 2, "offset": 0 , 'order':2}),
            
            
            dbc.Col(dcc.Dropdown(id='w_countries',
                         multi=True,
                         clearable=True,
                         disabled=False,
                         style={'display':True , 'background-color': 'rgb(218,165,27)' , 'font-weight': 'bold'},
                         value=['Portugal'],
                         placeholder='Select Nationality',
                         optionHeight=25,
                         searchable=True ,
                         persistence=True ,
                         persistence_type='session' ,
                         options=[{'label':c , 'value':c}
                                  for c in count_market1['country'].unique()]
                       
                           
),width={'size': 3, "offset":0 , 'order':4  }),
        dbc.Col(dcc.Dropdown( 
        id='w_status', 
        value='Check-Out', 
        options=[{'value': value, 'label': value} 
                    for value in count_market1['reservation_status'].unique()],
                 clearable=False ,
                  persistence=True ,
                  persistence_type='session' ,
                 optionHeight=25 ,
                 style={'color':'black' , 'background-color': 'rgb(218,165,27)' , 'font-weight': 'bold'}
        ),width={'size': 2, "offset":0 , 'order':3 })],className="mb-5 custom-dropdown"),    
     

         
     dbc.Row([
        dbc.Col(html.H3("Cancellations" , style={'font-size': '25px' , 'text-decoration': 'underline'}),
                        width={'size': 6, 'offset': 1 },
                        ),
        dbc.Col(html.H3("Market Segment" , style={'font-size': '25px' , 'text-decoration': 'underline'}),
                  width={'size':5 , 'offset': 0})
        ]),

    
     dbc.Row([
        dbc.Col(dcc.Graph(id='pie-chart', config={'scrollZoom':True , 'doubleClick':'reset',
                                                  'displayModeBar':False , 'watermark':True }) , width=6 , xs={'size':6}),
         
        dbc.Col(dcc.Graph(id='scatter' , config={'scrollZoom':True , 'doubleClick':'reset',
                                            'displayModeBar':False , 'watermark':True }) , width=6 , xs={'size':6} )
                    
            
        ]),
    
     dbc.Row([
        dbc.Col(html.H3("Reservations" , style={'font-size': '25px' , 'text-decoration': 'underline'}),
                        width={'size': 6, 'offset': 5 },
                        )]),
    
    dbc.Row([
        dbc.Col(dcc.Graph (id='map' , config={'scrollZoom':True , 'doubleClick':'reset',
                                            'displayModeBar':False , 'watermark':True }), width=8 , lg={'size':12})]),
        dbc.Row([
        dbc.Col(html.H3("Total Revenue" , style={'font-size': '25px' , 'text-decoration': 'underline'}),
                        width={'size': 6, 'offset': 1 },  
                        )]),
        dbc.Row([ 
        dbc.Col(dcc.Graph (id='bar_chart' , config={'scrollZoom':True , 'doubleClick':'reset',
                                                  'displayModeBar':False , 'watermark':True }))])
        

        
),id='mainContainer' , style={'display':'flex' , 'flex-direction':'column', 'color':'#000000' }, className='custom_container')







@app.callback(
    Output('pie-chart', 'figure'), 
    [Input('w_hoteltype', 'value')], 
    [Input('w_years', 'value')],
    [Input('w_month' , 'value')])

def generate_chart(w_hoteltype,w_years , w_month):

    if w_hoteltype=='All':
        cancelations = a1[(a1['Year'] == int(w_years)) & (a1['Month'] == w_month)]
    else:
        cancelations = a1[(a1['Year'] == int(w_years))  & (a1['hotel'] == w_hoteltype) & (a1['Month'] == w_month)] 


    fig = px.pie(cancelations, values='Total Cancelations R', names='country' ,
                color='country' ,
                color_discrete_map={
                "France": "#0d3172",
                "Portugal": "#5f9ea0",
                "United Kingdom": "#a8a8a8",
                "Germany":"#5f91ec",
                "Spain": "#d1dffa"})
    
    fig.update_layout({
        'plot_bgcolor': 'rgba(0, 0, 0, 0)',
        'paper_bgcolor': 'rgba(0, 0, 0, 0)'
})
               
    return fig

@app.callback(
    Output('scatter', 'figure'),
    [Input('w_hoteltype', 'value')],
    [Input('w_countries', 'value')], 
    [Input('w_years', 'value')],
    [Input('w_month' , 'value')],
    [Input('w_status' , 'value')])


def generate_bar_chart(w_hoteltype,w_countries ,w_years , w_month , w_status):
  
    if w_hoteltype=='All':
        a2 = count_market1[(count_market1['Year'] == int(w_years)) & (count_market1['country'].isin(w_countries))  & (count_market1['Month'] == w_month) & (count_market1['reservation_status'] == w_status) ]
        #print(a2)
    else:
        a2 = count_market1[(count_market1['Year'] == int(w_years))  & (count_market1['hotel'] == w_hoteltype)  & (count_market1['country'].isin(w_countries))  & (count_market1['Month'] == w_month)& (count_market1['reservation_status'] == w_status)] 
    

    barchart=px.scatter(a2, x='market_segment', y='count_market' , color='country' , hover_name='hotel' , size='count_market',
                       labels={'market_segment':'Market Segment' , 'count_market':'Quantity of markets'})
    
    barchart.update_layout({
        'plot_bgcolor': 'rgba(0, 0, 0, 0)',
        'paper_bgcolor': 'rgba(0, 0, 0, 0)'})
                
    return barchart
        

@app.callback(
    Output('map', 'figure'),
    [Input('w_hoteltype', 'value')],
    [Input('w_years', 'value')],
    [Input('w_month' , 'value')],
    [Input('w_status' , 'value')])


def generate_map(w_hoteltype, w_years , w_month , w_status):
    
    if w_hoteltype=='All':
        a3 = count_reserv1[(count_reserv1['Year'] == int(w_years))   & (count_reserv1['Month'] == w_month)  & (count_reserv1['reservation_status'] == w_status) ]
    else:
        a3 = count_reserv1[(count_reserv1['Year'] == int(w_years))  & (count_reserv1['hotel'] == w_hoteltype)  & (count_reserv1['Month'] == w_month) & (count_reserv1['reservation_status'] == w_status) ]
     
    

        
    fig = px.scatter_geo(a3, locations="country", color="country",
                     hover_name="Country", size="count_reserv",
                     
                     projection="natural earth")
    
    fig.update_layout({
        'plot_bgcolor': 'rgba(0, 0, 0, 0)',
        'paper_bgcolor': 'rgba(0, 0, 0, 0)'})
                
    return fig


@app.callback(
    Output('content-connections','children'),
    [Input('w_hoteltype', 'value')], 
    [Input('w_years', 'value')],
    [Input('w_month' , 'value')])

def generate_card_checkout(w_hoteltype ,w_years , w_month):

    dff_c = checkout.copy()
    
    if w_hoteltype=='All':
        dff_c = dff_c[(dff_c['Year'] == int(w_years))  & (dff_c['Month'] == w_month)]
        conctns_num = len(dff_c)   
    else:
        dff_c = dff_c[(dff_c['Year'] == int(w_years))  & (dff_c['hotel'] == w_hoteltype) & (dff_c['Month'] == w_month)]
        conctns_num = len(dff_c)

        
        
    return conctns_num
        
    
@app.callback(
    Output('content-connections1','children'),
    [Input('w_hoteltype', 'value')], 
    [Input('w_years', 'value')],
    [Input('w_month' , 'value')])

def generate_card_reservations(w_hoteltype ,w_years , w_month):
 
    dff_c = reservations.copy()
    
    if w_hoteltype=='All':
        dff_c = dff_c[(dff_c['Year'] == int(w_years))  & (dff_c['Month'] == w_month)]
        conctns_num1 = len(dff_c)   
    else:
        dff_c = dff_c[(dff_c['Year'] == int(w_years))  & (dff_c['hotel'] == w_hoteltype) & (dff_c['Month'] == w_month)]
        conctns_num1 = len(dff_c)

        
        
    return conctns_num1


@app.callback(
    Output('content-connections2','children'),
    [Input('w_hoteltype', 'value')], 
    [Input('w_years', 'value')],
    [Input('w_month' , 'value')])

def generate_card_canceled(w_hoteltype ,w_years , w_month):
 
    dff_c = canceled.copy()
    if w_hoteltype=='All':
        dff_c = dff_c[(dff_c['Year'] == int(w_years))  & (dff_c['Month'] == w_month)]
        conctns_num2 = len(dff_c) 
    else:
        dff_c = dff_c[(dff_c['Year'] == int(w_years))  & (dff_c['hotel'] == w_hoteltype) & (dff_c['Month'] == w_month)]
        conctns_num2 = len(dff_c)

        
        
    return conctns_num2

@app.callback(
    Output('content-connections3','children'),
    [Input('w_hoteltype', 'value')], 
    [Input('w_years', 'value')],
    [Input('w_month' , 'value')])

def generate_card_canceled(w_hoteltype ,w_years , w_month):
  
    dff_c = noshow.copy()
    
    if w_hoteltype=='All':
        dff_c = dff_c[(dff_c['Year'] == int(w_years))  & (dff_c['Month'] == w_month)]
        conctns_num3 = len(dff_c) 
    else:
        dff_c = dff_c[(dff_c['Year'] == int(w_years))  & (dff_c['hotel'] == w_hoteltype) & (dff_c['Month'] == w_month)]
        conctns_num3 = len(dff_c)

        
        
    return conctns_num3

@app.callback(
    Output('bar_chart', 'figure'), 
    [Input('w_hoteltype', 'value')], 
    [Input('w_years', 'value')],
    [Input('w_month' , 'value')])


def generate_scatter(w_hoteltype,w_years, w_month):
    
    adr1 = adr_new.copy()

    if w_hoteltype=='All':
        adr6 = adr1[(adr1['Year'] == int(w_years))  & (adr1['Month'] == w_month)]
        adr6 = adr6.sort_values('totaladr', ascending=False)

    else:
        adr6 = adr1[(adr1['Year'] == int(w_years))  & (adr1['hotel'] == w_hoteltype)  & (adr1['Month'] == w_month)]
        adr6 = adr6.sort_values('totaladr', ascending=False)
        
    barchart=px.bar(
        adr6 ,
        x='market_segment' , 
        y='totaladr',
        color=adr6['market_segment'] ,
        hover_name=adr6['hotel'],
        color_discrete_map={
                "Complementary": "rgb(166,118,29)",
                "Corporate": "#19D3F3",
                "Direct": "rgb(27,158,119)",
                "Groups":"#5f91ec",
                "Offline TA/TO": "rgb(218,165,27)" ,
                "Online TA": "rgb(47,138,196)" ,
                "Aviation":"rgb(47,138,196)"} , 
        labels={'market_segment':'Market Segment' , 'totaladr':'ADR'})
    
    barchart.update_layout({
        
        'plot_bgcolor': 'rgba(0, 0, 0, 0)',
        'paper_bgcolor': 'rgba(0, 0, 0, 0)'})

    return(barchart)


if __name__ == '__main__':
    app.run_server(debug=True , port=8052)
                
                
```
